```python
import matplotlib.pyplot as plt
import numpy as np

def plot_log(timelog, costlog, init_cost=None):
    # timelog = [2, 5, 10]
    timelog.append(timelog[-1]*1.2)
    # costlog = [10, 6, 3]
    num = 1000
    dt = timelog[-1]/(num-1)
    t = np.linspace(0, timelog[-1], num)
    # init_cost = costlog[0] + 3
    c = init_cost*np.ones(num)
    for i in range(num):
        t_curr = i*dt
        for idx, j in enumerate(timelog):
            if t_curr > j:
                c[i] = costlog[idx]
    k = int(timelog[0]/dt)
    t = t[k:]
    c = c[k:]
    plt.plot(t, c)

ts, cs = [27.2685, 88.2502, 181.6061, 489.3238], [27.2764, 22.5350, 21.1509, 20.9852]
tc, cc = [333], [20.9852]
plot_log(ts, cs, 15)
plot_log(tc, cc, 15)
# ts, cs = [11.660787105560303, 35.12951111793518, 92.5538969039917, 300.0587902069092], [10.374632164380044, 10.063362327159235, 9.980519614684615, 9.980519614684615]
# tc, cc = [58.56145405769348, 300], [8.88762639587116, 8.88762639587116]
# plot_log(ts, cs, 12)
# plot_log(tc, cc, 12)
plt.legend(["ana*", "a*"])
plt.title("cost vs time")
plt.ylabel("cost")
plt.xlabel("time/s")
plt.savefig('costvstime.png')
plt.show()
```
