```python
import matplotlib.pyplot as plt
import numpy as np
# plt.style.use('science')

def plot_log(timelog, costlog, init_cost=100, time_max=2000):
    timelog += [time_max]
    costlog += [costlog[-1]]
    timelog.append(timelog[-1]*1.2)
    num = 1000
    dt = timelog[-1]/(num-1)
    t = np.linspace(0, timelog[-1], num)
    # init_cost = costlog[0] + 3
    c = init_cost*np.ones(num)
    for i in range(num):
        t_curr = i*dt
        for idx, j in enumerate(timelog):
            if t_curr > j:
                c[i] = costlog[idx]
    k = int(timelog[0]/dt)
    t = t[k:]
    c = c[k:]
    plt.plot(t, c)
```
